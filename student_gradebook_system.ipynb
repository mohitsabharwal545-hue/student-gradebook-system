```python
## IMPORT LIBRARIES
```

```python
import numpy as np
import pandas as pd
```

```python
## CREATE STUDENT DATASET
```

```python
# Student Names
students = [
    "Mohit", "Rahul", "Aman", "Priya", "Neha",
    "Rohit", "Simran", "Karan", "Anjali", "Vikas"
]

# Subject Names (4 Assignments)
subjects = [
    "Mathematics",
    "Physics",
    "Computer Science",
    "English"
]

# Marks of Students
marks = np.array([
    [85, 90, 91, 87],
    [78, 82, 79, 81],
    [92, 95, 94, 96],
    [70, 75, 74, 73],
    [88, 86, 89, 91],
    [81, 83, 82, 85],
    [95, 97, 98, 99],
    [76, 79, 80, 78],
    [89, 91, 90, 93],
    [84, 85, 86, 87]
])

# Create Gradebook
df = pd.DataFrame(marks, index=students, columns=subjects)
print("========== STUDENT GRADEBOOK ==========\n")
print(df)
```

```python
# Student Statistics
```

```python
student_stats = pd.DataFrame({
    "Mean": df.mean(axis=1),
    "Maximum": df.max(axis=1),
    "Minimum": df.min(axis=1)
})

print("\n========== STUDENT STATISTICS ==========\n")
print(student_stats)
```

```python
# Subject Statistics
```

```python
subject_stats = pd.DataFrame({
    "Mean": df.mean(),
    "Maximum": df.max(),
    "Minimum": df.min()
})

print("\n========== SUBJECT STATISTICS ==========\n")
print(subject_stats)
```

```python
# Overall Statistics
```

```python
print("\n========== OVERALL CLASS STATISTICS ==========")
print("Overall Average Marks :", round(marks.mean(), 2))
print("Highest Marks         :", marks.max())
print("Lowest Marks          :", marks.min())
```

```python
# Top Performer
```

```python
student_stats["Average"] = student_stats["Mean"]
top_student = student_stats["Average"].idxmax()

print("\nTop Performer :", top_student)
print("Average Marks :", round(student_stats.loc[top_student, "Average"], 2))
```

```python
# Save Gradebook to CSV
```

```python
df.to_csv("students_grades.csv")
student_stats.to_csv("student_statistics.csv")
subject_stats.to_csv("subject_statistics.csv")
```
